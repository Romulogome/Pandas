```python
#Importando a Biblioteca
import pandas as pd
```

```python
#Trabalhando com arquivos EXCEL
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
#JUntando todos os arquivos EXCEL em um único arquivo para menor trabalho
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#Exibindo as linhas iniciais
df.head()
```

```python
#Exibindo as linhas finais
df.tail()
```

```python
#Pegando uma amostra dos dados
df.sample(5)
```

```python
#Verificando o tipo de dados
df.dtypes
```

```python
#Alterando um tipo de dado de uma coluna
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

```python
#Consultando valores nulos
df.isnull().sum()
```

```python
#Substituindo valores pela média
df["Vendas"].fillna(df["Vendas"].mean(), implace=True)
```

```python
df["Vendas"].mean()
```

```python
df.isnull().sum()
```

```python
df.sample(15)
```

```python
#Substituindo os valres por 0
df["Vendas"].fillna(0, imaplace=True)
```

```python
#Apagando as linhas com valores nulos
df.dropna(implace=True)
```

```python
#Apagando as linhas com valores nulos com base em 1 coluna
df.dropna(subset=["Vendas"], implace=True)
```

```python
#Remover as linhas com valores nulos em todas as colunas
df.dropna(how="all", implace=True)
```

```python
#Criando a coluna da receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
df["Receita/Vendas"] = df["Receita"]/df["Vendas"]
```

```python
df.head()
```

```python
#Encontrando a maior receita
df["Receita"].max()
```

```python
#Encontrando a menor receita
df["Receita"].min()
```

```python
#nlargest
df.nlargest(3, "Receita")
```

```python
#nsmallest
df.nsmallest(3, "Receita")
```

```python
#Agrupamento por cidades
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```
